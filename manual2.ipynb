```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mult = np.array([10, 80, 37, 17, 90, 31, 50, 20, 73, 89])
inhabitants = np.array([1, 6, 3, 1, 10, 2, 4, 2, 4, 8])
perc_contest = np.arange(40)
```

```python
fig, ax = plt.subplots()
for i in range(len(mult)):
    ax.plot(perc_contest, mult[i] / (inhabitants[i] + perc_contest), label=f'mult {mult[i]}')
ax.legend()
ax.set_xlabel('Percentage of contestants')
```
