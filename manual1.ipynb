```python
import numpy as np
```

```python
exchange_table = np.array([[1, 1.45, 0.52, 0.72],
                           [0.7, 1, 0.31, 0.48],
                           [1.95, 3.1, 1, 1.49],
                           [1.34, 1.98, 0.64, 1]])

labels = ["SB", "P", "SN", "SS"]
labels_map = {label: i for i, label in enumerate(labels)}

allpaths = []
allvals = []

def exchange(path, to_cur, val):
    from_cur = labels_map[path[-1]]
    val = val * exchange_table[from_cur][to_cur]
    path.append(labels[to_cur])
    return val, path

def exchange_all_options(val, path, depth):
    if depth == 1:
        allpaths.append(path)
        allvals.append(val)
        return val, path

    for i in range(len(exchange_table)):
        new_val, new_path = exchange(path.copy(), i, val)
        exchange_all_options(new_val, new_path, depth - 1)

exchange_all_options(1, ["SS"], 6)
print(allpaths)
print(allvals)
```

```python
valid_pathvals = [(x, allvals[i]) for i, x in enumerate(allpaths) if x[-1]=="SS"]
valid_path, valid_val = zip(*valid_pathvals)

max_arg = np.argmax(valid_val)

print(f"Max value: {valid_val[max_arg]}")
print(f"Path: {valid_path[max_arg]}")
```
